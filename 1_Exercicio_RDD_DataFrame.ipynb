```python
from pyspark.sql import SparkSession
from pyspark.sql import functions  as func
```

```python
spark = SparkSession.builder.appName("Spark Study").getOrCreate()
```

```python
client_data = spark.read.parquet("/content/Clientes.parquet")
```

```python
client_data.show()
```

```python
vendas_data = spark.read.parquet("/content/Vendas.parquet")
vendas_data.show()
```

```python
client_data.select("Cliente", "Estado", "Status").show()
```

```python
client_data.select("Cliente", "Estado", "Status") \
.where((func.col("Status")=="Platinum") | (func.col("Status")=="Gold")).show()
```

```python
client_data.join(vendas_data, client_data.ClienteID == vendas_data.ClienteID) \
.groupBy("Status") \
.agg(func.sum("Total").alias("TotalVendas")) \
.orderBy("TotalVendas",ascending=False).show()
```
