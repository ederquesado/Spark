```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as Func
```

```python
spark = SparkSession.builder.appName("Spark SQL").getOrCreate()
```

```python
client_data = spark.read.parquet("/content/Clientes.parquet")
itens_data = spark.read.parquet("/content/ItensVendas.parquet")
produtos_data = spark.read.parquet("/content/Produtos.parquet")
vendas_data = spark.read.parquet("/content/Vendas.parquet")
vendedores_data = spark.read.parquet("/content/Vendedores.parquet")

client_data.show()
itens_data.show()
produtos_data.show()
vendas_data.show()
vendedores_data.show()
```

```python
client_vendas_data = client_data.join(vendas_data, "ClienteID", "inner")
client_vendas_data.show()
```

```python
client_vendas_vendedores_data = client_vendas_data.join(vendedores_data, "VendedorID", "inner")
client_vendas_vendedores_data.show()
```

```python
itensV_prod_data = itens_data.join(produtos_data, "ProdutoID", "inner")
itensV_prod_data.show()
```

```python
all_data = client_vendas_vendedores_data.join(itensV_prod_data, "VendasID", "inner")
all_data.show()
```

```python

all_data.select(
    "Cliente",
    "Data",
    "Produto",
    "Vendedor",
    "ValorTotal"
).show()
```
