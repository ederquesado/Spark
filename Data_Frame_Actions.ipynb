```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as Func
from pyspark.sql.functions import col


spark = SparkSession.builder.appName("RDD Study").getOrCreate()
```

```python
from pyspark.sql.types import *
arqschema = "id INT, nome STRING, status STRING, cidade STRING, vendas INT, data STRING"
desp = spark.read.csv("despachantes.csv", header=False, schema=arqschema)
```

```python
#formato tabular
desp.show(1)
```

```python
#formato de lista
desp.take(1)
```

```python
#retorna todos dados como uma lista
desp.collect()
```

```python
#contagem
desp.count()
```

#trasnformações

```python
#padrão crescente
desp.orderBy("vendas").show()
```

```python
#decrescente
desp.orderBy(Func.col("vendas").desc()).show()
```

```python
#se quiser cidade dec e valor dec
desp.orderBy(Func.col("cidade").desc(),Func.col("vendas").desc()).show()
```

```python
from pyspark.sql.functions import sum

desp.groupBy("cidade").agg(sum("vendas")).show()
```

```python
#ordernar por vendas decrecente
desp.groupBy("cidade").agg(sum("vendas")).orderBy(Func.col("sum(vendas)").desc()).show()

```

```python
#filter
desp.filter(Func.col("nome") == "Deolinda Vilela").show()
```
