```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as Func
from pyspark.sql.functions import col


spark = SparkSession.builder.appName("RDD Study").getOrCreate()
```

```python
df1 = spark.createDataFrame([("Pedro",10),("Maria",20),("José",40)])
df1.show()
```

```python
#criar df com schema
schema = "Id INT, Nome STRING"
dados= [[1,"Pedro"],[2,"Maria"]]
df2 = spark.createDataFrame(dados,schema)
df2.show()
```

```python
#com transformação
from pyspark.sql.functions import sum
schema = "Produtos STRING, Vendas INT"
vendas = [["Caneta", 10],["Lapis", 20], ["Caneta", 40]]
df3 = spark.createDataFrame(vendas, schema)
df3.show()
```

```python
#selecionar colunas específicas
df3.select("Produtos").show()
```

```python
df3.select("Produtos","Vendas").show()
```

```python
df3.select("Vendas").show()
```

```python
df3.select("Produtos", "Vendas").where(col("Vendas")>20).show()
```

```python
from pyspark.sql.types import *
arqschema = "id INT, nome STRING, status STRING, cidade STRING, vendas INT, data STRING"
desp = spark.read.csv("despachantes.csv", header=False, schema=arqschema)
```

```python
desp.show()
```

```python
#outro exemplo, inferindo schema, usando load e informado tipo
desp2 = spark.read.load("despachantes.csv", format="csv", sep=",", inferSchema=True, header=False)
desp2.show()
```

```python
#comparando os schemas
desp.schema
```

```python
#comparando os schemas
desp2.schema
```

```python
#condição lógica com where
desp.select("id", "nome", "vendas").where(Func.col("vendas")>20).show()
```

```python
desp.select("id","nome","vendas").where((Func.col("vendas") > 20) & (Func.col("vendas") < 40)).show()
```

```python
novodf = desp.withColumnRenamed("nome","nomes")
novodf.columns
```

```python
from pyspark.sql.functions import *
desp3 = desp.withColumn("data2", to_timestamp(Func.col("data"),"yyyy-MM-dd"))
desp3.schema
```

```python
desp3.select(year("data")).show()
```

```python
desp3.select(year("data")).distinct().show()
```

```python
desp3.select("nome",year("data")).orderBy("nome").show()
```

```python
desp3.select("data").groupBy(year("data")).count().show()
```

```python
desp3.select(Func.sum("vendas")).show()
```
